```python
from eunjeon import Mecab

sentences = input()
m = Mecab()
m.morphs(sentences)

#기호를 나타내는 품사인 SF의 인덱스를 딴다
#SF가 온점일 때, 이전 SF의 인덱스 +1부터 온점인 SF까지의 문장을 출력한다.
```

```python
all_input_sentences = []
final_num = [] #문장 내 모든 마침표들의 위치

for i in range(len(m.morphs(sentences))) :
    if m.morphs(sentences)[i] == '.' :
        final_num.append(i)
    elif m.morphs(sentences)[i] == '?' :
        final_num.append(i)
    elif m.morphs(sentences)[i] == '!' :
        final_num.append(i)

final_num
```

```python
for j in range(len(final_num)):
    if j == 0 :
        all_input_sentences.append(m.morphs(sentences)[:final_num[j]+1])
    elif j > 0 :
        all_input_sentences.append(m.morphs(sentences)[final_num[j-1] + 1 : final_num[j] + 1])

for k in range(len(all_input_sentences)):
    print(all_input_sentences[k])

# dot_num = [] #문장 내 온점의 위치

# for k in range(len(m.morphs(sentences))):
#     if m.morphs(sentences)[k] == '.' :
#         dot_num.append(k)

# dot_num

# 나는 오늘 행궁동에서 커피를 마셨다. 사진도 여러장 찍었다.
```
